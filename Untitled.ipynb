```python
from cmdstanpy import CmdStanModel, cmdstan_path, install_cmdstan
import pickle
import warnings
```

```python
from inference.inference import hmc, meanfield_vi, map_opt, glickman

glicko_stan = './model/model.stan'
chess_data = './data/processed/data.pkl'
with open(chess_data, 'rb') as f:
    observed_data = pickle.load(f)
glicko_stan = CmdStanModel(stan_file=glicko_stan)
```

```python
glicko_map, score_pp_map = map_opt(glicko_stan, observed_data)
```

```python
from criticism.criticism import get_misclassification_error, get_posterior_predictions, get_binary_cross_entropy

glicko_map, score_pp_map = map_opt(
    glicko_stan,
    observed_data,
    iter=3500,
    tol_obj=1e-13,
    tol_rel_obj=1e3,
    tol_grad=1e-9,
    tol_rel_grad=1e8,
    tol_param=1e-9,
    history_size=4
)
```

```python
y_pred_MLE = get_posterior_predictions(
    model=glicko_map,
    quantity="score_ppd"
)

get_misclassification_error(y=observed_data["score_test"], y_pred=y_pred_MLE)
```
